# Metadatos XML con la iso 19139
Se asocian los metadatos al produto final a partir del siguiente Script el python

```python
import os
from arcpy import metadata as md
import arcpy
```

```python
folder = r"\\172.26.0.20\Elite_Sub_Geografia_Cartografia\Carto\Municipios"
for path, directories, files in os.walk(folder):
    for name in directories:
        if name.endswith(".gdb"):
            try:
                bbdd=os.path.join(path,name)
                print(bbdd)
                cut = bbdd.split("\\")
                xml = cut[-1].split(".")[0].split("_")
                xml.insert(1, "Metadato")
                xml = '\\'.join(cut[:-1])+'\\'+'_'.join(xml)+str(".xml")
                if arcpy.Exists(xml)==True:
                    tgt_item_md = md.Metadata(bbdd)
                    print(xml)
                    tgt_item_md.importMetadata(xml,"ISO19139_UNKNOWN")
                    tgt_item_md.synchronize("ALWAYS")
                    tgt_item_md.upgrade("ISO19139_UNKNOWN")
                    tgt_item_md.save()
            except:
                next 
```
