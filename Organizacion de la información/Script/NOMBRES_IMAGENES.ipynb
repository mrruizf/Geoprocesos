```python
import os
import time
import pandas as pd
```

```python
entrega = pd.read_excel(r'C:\Users\marlon.ruiz\Downloads\Entrega_20221108.xlsx')
```

```python
entrega = entrega[entrega['Producto']=='MODELO DIGITAL DE TERRENO']
```

```python
Listaraster=[]
Listaraster1 = []
divipola = []
modelos = []
for path, directories, files in os.walk(r"Z:\87RVOficializacion\1PExternos\C_C"):
    for name in files:
        if (name.endswith(".tif") or name.endswith(".TIF")) and (name.startswith("MDT") or name.startswith("DEM")):
            try:
                Rutaraster=os.path.join(path,name)
                Listaraster.append(Rutaraster)
                divipola.append(name.split("_")[1])
                modelos.append(name)
                Listaraster1.append(time.strftime('%d/%m/%Y', time.gmtime(os.path.getmtime(os.path.join(Rutaraster)))))
                print("la ruta",Rutaraster,"\n Divipola",name.split("_")[1])
            except FileNotFoundError:
                next

entregaDF = pd.DataFrame(list(zip(divipola,modelos,Listaraster1,Listaraster)),columns=['DANE','PRODUCTOS','FECHAS','PATH'])
```

```python
entrega['DIVIPOLA']= entrega['DIVIPOLA'].values.astype(str)
```

```python
union = entrega.merge(entregaDF, how='left', left_on='DIVIPOLA', right_on='DANE')
```

## Exportando el excel para control

```python
union.to_excel('Entrega_20221108.xlsx', sheet_name='Entregas')
```

```python
union = union[union['PRODUCTOS'].duplicated(False)!=True].reset_index()
```

## Creando las carpetas y copiando las imagenes a la ruta de entregas

```python
import arcpy
```

```python
import shutil
shutil.rmtree(r"\\172.26.0.20\Elite_Sub_Geografia_Cartografia\MD\Entregas\ENTREGAS 2022\MDT10")
```

```python
destino = r"\\172.26.0.20\Elite_Sub_Geografia_Cartografia\MD\Entregas\ENTREGAS 2022"
for index, row in union.iterrows():
    folder = os.path.join(destino,row['PRODUCTOS'].split(".")[0])
    if not os.path.exists(folder):
        try:
            os.makedirs(folder)
            with arcpy.EnvManager(compression="NONE"):
                arcpy.management.CopyRaster(row['PATH'],os.path.join(folder,row['PRODUCTOS']), '', None, "-3,402823e+38", "NONE", "NONE", '', "NONE", "NONE", "TIFF", "NONE", "CURRENT_SLICE", "NO_TRANSPOSE")
        except arcpy.ExecuteError:
            msgs = arcpy.GetMessages(2)
            print(u"Error de Arcpy :\n {}".format(msgs))
            pass
    else:
        print('Ya existe la Carpeta')
```
