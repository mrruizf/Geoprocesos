```python
import os
import pandas as pd
```

```python
folder = r"E:\BORRAR"
```

```python
lista_path = []
for path, directories, files in os.walk(folder):
    for name in directories:
        if name.endswith(".gdb"):
            try:
                #print(os.path.join(path,name)+"\t--\n"+name[4:])
                lista_path.append(os.path.join(path,name))
            except:
                next
```

```python
new_list = []
for i in lista_path:
    #print(i.replace('gdb_fc','gdb'))
    new_list.append(i.replace('BORRAR','ENTREGA'))
```

```python
new_list
```

```python
df = pd.DataFrame(new_list)
df[['E', 'root','Depto','Folder','gdb']] = df[0].str.split('\\', expand=True)
df['Path'] = df['E']+"\\"+df['root']+"\\"+df['Depto']+"\\"+df['Folder']
df
```

```python
for i in range(len(df)):
    path = str(df.iloc[i]['E'])+"\\"+str(df.iloc[i]['root'])+"\\"+str(df.iloc[i]['Depto'])
    if os.path.isdir(path)==False:
        os.makedirs(path)
        print(path)
    path2 = path+"\\"+str(df.iloc[i]['Folder'])
    if os.path.isdir(path2)==False:
        os.makedirs(path2)
        print(path2)
```
