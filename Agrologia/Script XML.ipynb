```python
def register_all_namespaces(filename):
    namespaces = dict([node for _, node in ET.iterparse(filename, events=['start-ns'])])
    for ns in namespaces:
        ET.register_namespace(ns, namespaces[ns])

ET.register_namespace("", "http://www.isotc211.org/2005/gmd")
ET.register_namespace("gco", "http://www.isotc211.org/2005/gco")
ET.register_namespace("gts", "http://www.isotc211.org/2005/gts")
ET.register_namespace("srv", "http://www.isotc211.org/2005/srv")
ET.register_namespace("gml", "http://www.opengis.net/gml/3.2")
ET.register_namespace("gmx", "http://www.isotc211.org/2005/gmx")
ET.register_namespace("xlink", "http://www.w3.org/1999/xlink")
ET.register_namespace("xsi", "http://www.w3.org/2001/XMLSchema-instance")
for i in range(len(data)):
    #register_all_namespaces(r'D:\gdb\0 Plantilla\AHT_DDMMM_Metadato_Dpt_Mun_Año.xml')
    tree = ET.parse(r'D:\gdb\0 Plantilla\AHT_DDMMM_Metadato_Dpt_Mun_Año.xml')
    root = tree.getroot() 
    try:
        j = 0
        k = 0
        for elem in root.iter('{http://www.isotc211.org/2005/gco}CharacterString'):
            if j in [10,11,22,27,28,48]:
                if j in [10,11]:
                    elem.text = data.iloc[i]['folder']
                elif j == 22:
                    string = elem.text
                    x = string.find("(MPIO)")
                    y = string.find("(ANIO)")
                    elem.text = string[0:x]+data.iloc[i]['MpNombre']+string[x+6:y]+data.iloc[i]['anios']+string[y+7:]
                elif j == 27:
                    string = elem.text
                    x = string.find("(MPIO)")
                    elem.text = string[0:x]+str(data.iloc[i]['MpNombre'])
                elif j == 28:
                    string = elem.text
                    x = string.find("(DEPTO)")
                    elem.text = string[0:x]+str(data.iloc[i]['Depto'])
                elif j == 48:
                    string = elem.text
                    x = string.find("(MPIO)")
                    y = string.find("(DEPTO)")
                    elem.text = string[0:x]+data.iloc[i]['MpNombre']+string[x+6:y]+data.iloc[i]['Depto']+string[y+7:]
            j+=1
        for elem in root.iter('{http://www.isotc211.org/2005/gco}Decimal'):
            if k in [0,4]:
                elem.text = str(data.iloc[i]['EXT_MIN_X'])
            elif k in [1,5]:
                elem.text = str(data.iloc[i]['EXT_MAX_X'])
            elif k in [2,6]:
                elem.text = str(data.iloc[i]['EXT_MIN_Y'])
            elif k in [3,7]:
                elem.text = str(data.iloc[i]['EXT_MAX_Y'])
            k+=1
        #ET.dump(tree)
        
        """ET.register_namespace("", "http://www.isotc211.org/2005/gmd")
        ET.register_namespace("gco", "http://www.isotc211.org/2005/gco")
        ET.register_namespace("gts", "http://www.isotc211.org/2005/gts")
        ET.register_namespace("srv", "http://www.isotc211.org/2005/srv")
        ET.register_namespace("gml", "http://www.opengis.net/gml/3.2")
        ET.register_namespace("gmx", "http://www.isotc211.org/2005/gmx")
        ET.register_namespace("xlink", "http://www.w3.org/1999/xlink")
        ET.register_namespace("xsi", "http://www.w3.org/2001/XMLSchema-instance")"""
        tree.write(str(data.iloc[i]['xml']), encoding="utf-8",xml_declaration=True)
        print("\t\tFinalizado")
    except Exception:
        print("No realizado")
        pass
```
