```python
from arcgis.gis import GIS 
import xml.dom.minidom as md
import time
import os
```

```python
Listaraster=[]
with open('salida_georre_70k_3.txt', 'w') as f:
    for path, directories, files in os.walk(r"D:\Imagenes Analogas"):
        for name in files:
            if name.endswith(".tif"):
                try:
                    #print(path)
                    #print(os.path.join(path,name))
                    Rutaraster=os.path.join(path,name)
                    #print("la ruta",Rutaraster)
                    Listaraster.append([time.strftime('%d/%m/%Y', time.gmtime(os.path.getmtime(os.path.join(Rutaraster)))),Rutaraster])
                    f.write("%s\n" % Rutaraster)
                    print(time.strftime('%d/%m/%Y', time.gmtime(os.path.getmtime(os.path.join(Rutaraster))))," nombre de la imagen ",name)
                except FileNotFoundError:
                    next                
```
