```python
import arcpy
import os
import pandas as pd

df = pd.read_excel(r"C:\Users\marlon.ruiz\Downloads\Orientaciones_70K.xlsx")

df_imagenes = pd.read_excel(r"C:\Users\marlon.ruiz\Downloads\Coords_imagenes_70K.xlsx")

data_union = pd.merge(df, df_imagenes, left_on=['Ruta','Corner'], right_on=['Ruta_Imagen','Corner_Imagen'])#, how='left').drop('id1', axis=1)
data_union = data_union[["Image","Ruta","Este","Norte","Corner","Ruta_Imagen","Este_Imagen","Norte_Imagen","Corner_Imagen"]]
```

```python
df_espacio0 = pd.read_table(r"C:\Users\marlon.ruiz\Downloads\salida_georre_70k_1.txt")
df_espacio0 = df_espacio0['Ruta'].str.split('\\', expand=True)
df_espacio1 = pd.read_table(r"C:\Users\marlon.ruiz\Downloads\salida_georre_70k_2.txt")
df_espacio1 = df_espacio1['Ruta'].str.split('\\', expand=True)
df_espacio2 = pd.read_table(r"C:\Users\marlon.ruiz\Downloads\salida_georre_70k_3.txt")
df_espacio2 = df_espacio2['Ruta'].str.split('\\', expand=True)
df_espacio3 = pd.read_table(r"C:\Users\marlon.ruiz\Downloads\salida_georre_70k_4.txt")
df_espacio3 = df_espacio3['Ruta'].str.split('\\', expand=True)

df_espacio = pd.concat([df_espacio0, df_espacio1, df_espacio2,df_espacio3])
df_espacio.columns = ['E', 'folder', 'folder1', 'imagen']
df_espacio["Ruta"] = r"\\172.26.0.20\Cartografia\Vuelos\Analogas\\"[:-1] + df_espacio["folder1"] + "\\" + df_espacio["imagen"]
```

```python
data_union = pd.merge(data_union,df_espacio,on='Ruta',how='left').loc[pd.merge(data_union,df_espacio,on='Ruta',how='left')['E'] != 'E:'] # 'D:'
#data_union
#len(set(data_union[["Image","Ruta","Este","Norte","Corner","Ruta_Imagen","Este_Imagen","Norte_Imagen","Corner_Imagen"]]["Image"]))
data_union = data_union.reset_index()
```

```python
transformation_type = "POLYORDER1"
resampling_type = "NEAREST"
for i in range(len(list(data_union['Image']))):#range(len(list(data_union['image'])))
    try:
        if str(data_union['Image'][i]) == str(data_union['Image'][i+1]) and str(data_union['Image'][i]) == str(data_union['Image'][i+2]) and str(data_union['Image'][i]) == str(data_union['Image'][i+3]):
            in_raster = data_union['Ruta'][i]
            source_control_points = "'"+str(data_union['Este_Imagen'][i]).replace(".", ",")+' '+str(data_union['Norte_Imagen'][i]).replace(".", ",")+"';'"+str(data_union['Este_Imagen'][i+1]).replace(".", ",")+' '+str(data_union['Norte_Imagen'][i+1]).replace(".", ",")+"';'"+str(data_union['Este_Imagen'][i+2]).replace(".", ",")+' '+str(data_union['Norte_Imagen'][i+2]).replace(".", ",")+"';'"+str(data_union['Este_Imagen'][i+3]).replace(".", ",")+' '+str(data_union['Norte_Imagen'][i+3]).replace(".", ",")+"'"
            target_control_points = "'"+str(data_union['Este'][i]).replace(".", ",")+' '+str(data_union['Norte'][i]).replace(".", ",")+"';'"+str(data_union['Este'][i+1]).replace(".", ",")+' '+str(data_union['Norte'][i+1]).replace(".", ",")+"';'"+str(data_union['Este'][i+2]).replace(".", ",")+' '+str(data_union['Norte'][i+2]).replace(".", ",")+"';'"+str(data_union['Este'][i+3]).replace(".", ",")+' '+str(data_union['Norte'][i+3]).replace(".", ",")+"'"
            path = r'E:\08 Junio'+"\\"+in_raster.split("\\")[-2]
            #print(in_raster)
            #print(source_control_points)
            #print(target_control_points)
            #print(path)
            if os.path.isdir(path)==False:
                #print("Se crea la carpeta de salida "+path)
                os.makedirs(path)
            else:
                print('Ya existe la carpeta de salida '+path)
            out_raster = path+"\\"+in_raster.split("\\")[-1]
            #print("Imagen de salida "+out_raster)
            with arcpy.EnvManager(compression="NONE", parallelProcessingFactor="1"):
                arcpy.management.Warp(in_raster,source_control_points,target_control_points, out_raster,transformation_type,resampling_type)
                print("Realizado")
    except Exception:
        pass
```
